### **&#167; Task:**  **1.0 Write a function to the find the largest prime from a list of arbitrary size.**


This should be very straightforward, but remember:

1. you will need to first check if a number is a candidate to be a 
   prime number.  Use the `is_prime()` function provided and this can be easily done.
   Also remember
   the following fact:  _there are no even prime numbers_, or
   alternatively, if a number is divisible by 2, it _cannot_ be prime,
             
2. _keep track_ of the largest prime that you have seen, thus
  a loop will be your best tool along with a variable, say `max_prime` to store
  the largest seen so far,

3. make sure that if you see a new prime check it against your `max_prime`,

2. don't overthink it, you only need one loop and some conditional `if-else` logic to
  check the current number in the loop,
  
4. your function should be named `find_largest_prime`.

````python
## is_prime() function to use 
##   in your solution

def is_prime(n):
    import math
    if n <= 1:
        return False
    else:
        is_prime = True
        for i in range(2, int(math.sqrt(n)) + 1):
            if n % i == 0: #Checks if the remainder equals to 0
                is_prime = False
                break
        return is_prime
````

_function_ **NAME**: `find_largest_prime`


_function_ **INPUT**

  - a Python list.

_function_ **OUTPUT**

  - `None` if there are no primes in the list,
  - the largest prime number in the list.

_Code Example_:
```python
  input_list = [1, 2, 3, 4, 5]
  largest_prime(input_list)
```

_Output_:
```python
  5
```

````python
### YOUR SOLUTION

def find_largest_prime(numbers):
    max_prime = 0
    
    for num in numbers:         
        if is_prime(num):
            if num > max_prime:
                max_prime = num
                
    return max_prime

# Verifying the function with examples given above
input_list0 = [1, 2, 3, 4, 5]
print(find_largest_prime(input_list0))

input_list1 =[834, 937, 1065, 1124, 1137, 1256, 1380, 1422, 1155, 1287, 1365]
print(find_largest_prime(input_list1))
````

````python

# YOUR SOLUTION GOES HERE

def find_words(s, w_size):
    # implementation here
    word=[]
    words= s.split()
    
    for w in words:
        if len(w) == w_size:
            word.append(w)
    return word

# Cross verification if it gives the words when the string size is changed. Randomly trying for 5 words and 7 words
input_string = mystery_function()
words5 = find_words(input_string, 5)
print(words5)

# For seven words
words7 = find_words(input_string, 7)
print(words7)
````

_function_ **OUTPUT**:
  
  - a string containing random words of the size of each number in `size_lst`
  
_Code Example_:



````python
# YOUR SOLUTION GOES HERE TO 

import random

def generate_sentence(size_lst, w_map):
    sentences = []
    for size in size_lst:
        if size in w_map and w_map[size]:
            word = random.choice(w_map[size])
            sentences.append(word)
        else:
            sentences.append("...")
            
    return " ".join(sentences).strip()


# Cross verification of the above function 
size_lst = [3,6,3,5,4,9]
print(generate_sentence(size_lst, d))
````
